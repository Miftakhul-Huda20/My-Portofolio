**Demographic Details**

```python
plt.figure(figsize=(18, 6))

# Ethnicity
plt.subplot(131)
sns.countplot(x='Ethnicity', data=df, palette='Set3')
plt.title('Ethnicity Distribution')
plt.xlabel('Ethnicity')
plt.ylabel('Count')
plt.xticks(ticks=[0, 1, 2, 3], labels=['Caucasian', 'African American', 'Asian', 'Other'])

# Socioeconomic Status
plt.subplot(132)
sns.countplot(x='SocioeconomicStatus', data=df, palette='Set2')
plt.title('Socioeconomic Status Distribution')
plt.xlabel('Socioeconomic Status')
plt.ylabel('Count')
plt.xticks(ticks=[0, 1, 2], labels=['Low', 'Middle', 'High'])

# Education Level
plt.subplot(133)
sns.countplot(x='EducationLevel', data=df, palette='Set1')
plt.title('Education Level Distribution')
plt.xlabel('Education Level')
plt.ylabel('Count')
plt.xticks(ticks=[0, 1, 2, 3], labels=['None', 'High School', "Bachelor's", 'Higher'])

plt.tight_layout()
plt.show()
```

```python
ethnicity_map = {0: 'Caucasian', 1: 'African American', 2: 'Asian', 3: 'Other'}
ses_map = {0: 'Low', 1: 'Middle', 2: 'High'}
edu_map = {0: 'None', 1: 'High School', 2: "Bachelor's", 3: 'Higher'}

df['Ethnicity'] = df['Ethnicity'].map(ethnicity_map)
df['SocioeconomicStatus'] = df['SocioeconomicStatus'].map(ses_map)
df['EducationLevel'] = df['EducationLevel'].map(edu_map)

# Create a cross-tabulation (pivot table) of the data
cross_tab = pd.crosstab(index=[df['Ethnicity'], df['SocioeconomicStatus']],
                        columns=df['EducationLevel'], normalize='index')

# Plotting
plt.figure(figsize=(12, 8))
sns.heatmap(cross_tab, annot=True, cmap='Blues', cbar=True, fmt='.2f')
plt.title('Interrelation between Ethnicity, Socioeconomic Status, and Education Level')
plt.xlabel('Education Level')
plt.ylabel('Ethnicity and Socioeconomic Status')
plt.show()
```
